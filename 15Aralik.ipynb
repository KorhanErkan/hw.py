```python
import random
from collections import Counter

yemekler = ["Hamburger", "Pizza", "Lahmacun"]
sayi_liste = list()
yemek_liste = list()
for i in range(365):
  sayi_liste.append(random.randint(1, 3))

for i in range(365):
  if sayi_liste[i] == 1:
    yemek_liste.append("Hamburger")
  elif sayi_liste[i] == 2:
    yemek_liste.append("Pizza")
  else:
    yemek_liste.append("Lahmacun")


olasiliklar = [0]*9

for i,j in zip(range(365),range(1,365)):
  if yemek_liste[i] == "Hamburger":
    if yemek_liste[j] == "Hamburger":
      olasiliklar[0] +=1
    elif yemek_liste[j] == "Pizza":
      olasiliklar[1] +=1
    else:
      olasiliklar[2] +=1
  
  elif yemek_liste[i] == "Pizza":
    if yemek_liste[j] == "Pizza":
      olasiliklar[3] +=1
    elif yemek_liste[j] == "Hamburger":
      olasiliklar[4] +=1
    else:
      olasiliklar[5] +=1

  else:
    if yemek_liste[j] == "Lahmacun":
      olasiliklar[6] +=1
    elif yemek_liste[j] == "Pizza":
      olasiliklar[7] +=1
    else:
      olasiliklar[8] +=1



sayilar = dict(Counter(yemek_liste))
print(sayilar)


print(olasiliklar[0]/sayilar["Hamburger"])
print(olasiliklar[1]/sayilar["Hamburger"])
print(olasiliklar[2]/sayilar["Hamburger"])
print(olasiliklar[3]/sayilar["Lahmacun"])
print(olasiliklar[4]/sayilar["Lahmacun"])
print(olasiliklar[5]/sayilar["Lahmacun"])
print(olasiliklar[6]/sayilar["Pizza"])
print(olasiliklar[7]/sayilar["Pizza"])
print(olasiliklar[8]/sayilar["Pizza"])
```

```python
durum = list()
```

```python
for i,j in zip(range(365*2),range(1,(365*2))):
  if altinfiyat[i] == altinfiyat[j]:
    durum.append("Sabit")
  elif altinfiyat[i] > altinfiyat[j]:
    durum.append("Azalmis")
  else:
    durum.append("Artmis")
```

```python
for i,j in zip(range(len(durum)),range(1,len(durum))):
  if durum[i] == "Sabit":
    if durum[j] == "Artmis":
      olasiliklar[0] +=1
    elif durum[j] == "Azalmis":
      olasiliklar[1] +=1
    else:
      olasiliklar[2] +=1
  elif durum[i] == "Artmis":
    if durum[j] == "Artmis":
      olasiliklar[3] +=1
    elif durum[j] == "Azalmis":
      olasiliklar[4] +=1
    else:
      olasiliklar[5] +=1
  else:
    if durum[j] == "Sabit":
      olasiliklar[6] +=1
    elif durum[j] == "Artmıs":
      olasiliklar[7] +=1
    else:
      olasiliklar[8] +=1

sayilar = dict(Counter(durum))
```

```python
text = "sometimes it happens that the most insane thought, the most impossible conception, will become so fixed in ones head that at length one believes the thought or the conception to be reality. Moreover, if with the thought or the conception there is combined a strong, a passionate, desire, one will come to look upon the said thought or conception as something fated, inevitable, and foreordained—something bound to happen."
```

```python
import nltk
import numpy as np
import pandas as pd
nltk.download('punkt')
import string
     
text = text.lower()    
len(text.split(". "))



a = nltk.word_tokenize(text)
     

b = list()
for i in a:
  if i not in string.punctuation:
    b.append(i)
  else:
    pass
a = b
     

b = list()
for i in a:
  if (i == "``") or (i.startswith("''")):
    pass
  elif i.startswith("-"):
    b.append(i[1:])
  else:
    b.append(i)
a = b
     

arr = np.unique(np.array(a))
     

my_df = pd.DataFrame(data=arr,columns=["words"])
my_df2 = pd.DataFrame(columns=arr)
my_df = my_df.set_index("words")
my_df = pd.concat([my_df,my_df2])
my_df = my_df.replace(np.nan,0)
     

my_df
```
